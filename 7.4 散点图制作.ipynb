```python
import numpy as np
import pandas as pd
from pylab import *
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel('均值与标准差.xlsx',encoding = 'gbk')
```

```python
data.head()
```

```python
data1 = data.drop(['按季节分类','直接分类','两阶聚类'],axis= 1)
data2 = data.drop(['直接分类','按七项分类','两阶聚类'],axis= 1)
data3 = data.drop(['按季节分类','按七项分类','两阶聚类'],axis= 1)
data4 = data.drop(['按季节分类','直接分类','按七项分类'],axis= 1)
```

```python
data10 = data1[data1['按七项分类'] == 0]
data11 = data1[data1['按七项分类'] == 1]
data12 = data1[data1['按七项分类'] == 2]
data13 = data1[data1['按七项分类'] == 3]
```

```python
# from matplotlib.font_manager import FontProperties #字体管理器
# #设置汉字格式
# from matplotlib import font_manager
# font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=20)
# plt.rcParams['font.sans-serif']=['SimHei']
# plt.rcParams['axes.unicode_minus']=False
# my_font = font_manager.FontProperties(fname='msyh.ttc')
plt.rc('font',family='Times New Roman') 
plt.figure(figsize=(9, 9), dpi=80)
plt.grid(alpha=0.3, linestyle='-.')
y = data10['标准差']
x = data10['均值']
plt.scatter(x,y,color='r',s=20, alpha=.7)
a = data11['标准差']
b = data11['均值']
plt.scatter(b,a,color='b',s=20, alpha=.7)
c = data12['标准差']
d = data12['均值']
plt.scatter(d,c,color='g',s=20, alpha=.7)
e = data13['标准差']
f = data13['均值']
plt.scatter(f,e,color='y',s=20, alpha=.7)
plt.xlabel('Mean electricity consumption/℃', fontsize=19)
plt.ylabel('Standard deviation of electricity consumption/kWh', fontsize=19)
plt.axis([0, 14, 0, 14],fontsize=17)
plt.legend(['Red spot:Cluster 1','Blue spot:Cluster 2','Green spot:Cluster 3','Yellow spot:Cluster 4'],loc = 'upper left',fontsize=15)
plt.savefig('按七项分类散点图.svg',format='svg')
plt.show()
```
