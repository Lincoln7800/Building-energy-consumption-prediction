```python
from matplotlib.font_manager import FontProperties #字体管理器
#设置汉字格式
from matplotlib import font_manager
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=20)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
my_font = font_manager.FontProperties(fname='msyh.ttc')
plt.figure(figsize=(10, 10), dpi=80)
plt.grid(alpha=0.3, linestyle='-.')
y = dataset3['consumption']
x = dataset3['temperature']
p1 = np.poly1d(np.polyfit(x,y,1))
p2 = np.poly1d(np.polyfit(x,y,2))
p3 = np.poly1d(np.polyfit(x,y,3))
xp = np.linspace(-10,30,104)#数据共有104条，0-30是用电量的范围
plt.scatter(x,y,color='r',s=20, alpha=.7)
plt.xlabel('温度/℃', fontproperties=font)
plt.ylabel('用电量/kWh', fontproperties=font)
plt.axis([-10, 30, 2, 20],fontsize=15)
plt.plot(xp,p1(xp),c='g')
plt.plot(xp,p2(xp),c='y')
plt.plot(xp,p3(xp),c='b')
matplotlib.rcParams.update({'font.size': 13})
plt.legend(['绿线：线性拟合','黄线：二次拟合','蓝线：三次拟合'],loc = 'upper right',fontsize=20)
# plt.title('偏好2拟合曲线', fontproperties=font)
plt.savefig('偏好3拟合曲线.svg',format='svg')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_2 = r2_score(y,p2(x))
r2_1 = r2_score(y,p1(x))
r2_3 = r2_score(y,p3(x))
r2_2,r2_1,r2_3
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y,p1(x))
MSE
RMSE = np.sqrt(MSE)
mean = np.mean(y)
print('CV-RMSE', RMSE/mean)
```
